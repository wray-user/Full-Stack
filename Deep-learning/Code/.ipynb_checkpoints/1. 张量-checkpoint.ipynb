## 2 01 
- torch.ones  torch.ones_like 1
- torch.zeros  torch.zeros_like 0
- torch.full  torch.full_like 



zeros(), full()

```python
import torch
import numpy as np
```

```python
import torch
# scene 1torch.ones  torch.ones_like
t1 = torch.ones(2, 3)   # 231
print(f't1:{t1}, type:{type(t1)}')
print('-'*30)

# t2: 32
t2 = torch.tensor([[1, 2],[3, 4], [5, 6]])
print(f't2:{t2}, type:{type(t2)}')
print('-'*30)

# t3 -> t21
t3 = torch.ones_like(t2)
print(f't3:{t3}, type:{type(t3)}')     # 32  -> 1
```

```python
# scene 2torch.zeros  torch.zeros_like
t1 = torch.zeros(2, 3)   # 230
print(f't1:{t1}, type:{type(t1)}')
print('-'*30)

# t2: 32
t2 = torch.tensor([[1, 2],[3, 4], [5, 6]])
print(f't2:{t2}, type:{type(t2)}')
print('-'*30)

# t3 -> t20
t3 = torch.zeros_like(t2)
print(f't3:{t3}, type:{type(t3)}')     # 32  -> 0
```

```python
# scene 3torch.full  torch.full_like 
t1 = torch.full(size = (2, 3), fill_value=255)   # 23255
print(f't1:{t1}, type:{type(t1)}')
print('-'*30)

# t2: 32
t2 = torch.tensor([[1, 2],[3, 4], [5, 6]])
print(f't2:{t2}, type:{type(t2)}')
print('-'*30)

# t3 -> t2255
t3 = torch.full_like(t2, fill_value=255)
print(f't3:{t3}, type:{type(t3)}')     # 32  -> 255
```

```python
import torch
```

```python
# 1. 
def dm04():
    # scene 1:  
    t1 = torch.arange(1, 10, 2)        # 1  2  3
    print(f't1:{t1}, type:{type(t1)}')
    print('-'*30)

    # scene 2:   -> 
    t2 = torch.linspace(1, 10, 5)     # 1  2  3
    print(f't2:{t2}, type:{type(t2)}')
    print('-'*30)
    
```

```python
dm04()
```

```python
# 2. 
def dm05():
    # step1: 
    # torch.initial_see()  # 
    torch.manual_seed(3)

    # step: 
    # scene 1:  (0, 1) 
    t1 = torch.rand(size=(2,3))
    print(f't1:{t1}, type:{type(t1)}')
    print('-'*30)
    
    # scene 2: 
    t2 = torch.randn(size=(2,3))     # 1  2  3
    print(f't2:{t2}, type:{type(t2)}')
    print('-'*30)

    # scene 3: 
    t3 = torch.randint(1, 10, size=(2,3))     # 1  2  3
    print(f't2:{t3}, type:{type(t3)}')
    print('-'*30)
```

```python
dm05()
```

```python
# 1.  -> numpy 
def dm06(): 
     # 1. 
    t1 = torch.tensor([1, 2, 3, 4, 5])
    print(f't1:{t1}, type:{type(t1)}')

    # 2.  -> numpy 
    # n1 = t1.numpy()  #  
    n1 = t1.numpy().copy()     # 
    print(f'n1:{n1}, type:{type(n1)}')

    # 3. 
    n1[0] = 100
    print(f'n1:{n1}')
    print(f't1:{t1}')
```

```python
dm06()
```

```python
def dm07():
    # 1.  numpy 
    n1 = np.array([11, 22, 33])
    print(f'n1:{n1}, type:{type(n1)}')

    # 2. numpyt 
    # t1 = torch.from_numpy(n1).type(torch.float32)  #  + 
    t1 = torch.from_numpy(n1)
    print(f't1:{t1}, type:{type(1)}')     # 

    t2 = torch.tensor(n1)    # 
    print(f't2:{t2}, type:{type(t2)}')  

    # 3. 
    n1[0] = 100 
    print(f'n1:{n1}') 
    print(f't1:{t1}') 
    print(f't2:{t2}') 
    
```

```python
dm07()
```
